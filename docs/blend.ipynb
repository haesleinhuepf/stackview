# Blending images


```python
import stackview
from skimage.io import imread
from skimage.measure import label
```

```python
image = imread("data/blobs.tif")
```

```python
label_image = label(image > 128)
```

```python
stackview.blend(image, label_image)
```

```python
stackview.animate_blend(image, label_image)
```
