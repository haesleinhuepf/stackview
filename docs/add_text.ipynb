# Adding text on images

```python
import numpy as np
import stackview
from skimage.io import imread
from skimage.filters import gaussian
import matplotlib.pyplot as plt
```

```python
image = imread('data/blobs.tif')
```

We'll apply Gaussian blur with sigma values from 1 to 10 and store the results in lists.

```python
# Initialize lists to store results
blurred_images = []
sigma_labels = []

# Apply Gaussian blur with sigma values from 1 to 10
for sigma in range(1, 11):
    blurred_image = gaussian(image, sigma=sigma, preserve_range=True)
    blurred_images.append(blurred_image)
    sigma_labels.append(f"sigma={sigma}")

print(f"Generated {len(blurred_images)} blurred images")
print(f"Sigma labels: {sigma_labels}")
```

First, we add text to images and show them afterwards.

```python
images = stackview.add_text(blurred_images, sigma_labels, position="bottom-left")
images[0].shape
```

```python
stackview.animate(images)
```
